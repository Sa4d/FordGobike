# Subscribers and customers percentage

#### In the following section we found a graph that gives us a clear describtion of the difference between the subscribers and the customers 
#### where we have 163,544 subscribers which is 89% of the users 
#### and we have 19,868 customers and it's 11% of the users 

```python
df['user_type'].value_counts(normalize=True).plot(kind='bar',figsize=(10,6))
plt.title('The percentge of the user types')
plt.xlabel('User Type')
plt.ylabel('The Percentage of the user')
plt.legend() ## if needed
plt.grid();
```

# Who uses the bikes the longest time?

#### In the next graph we can notice something, although we have only 11% customers but they are the ones who uses the bikes the longest time, and they have higher time duration spent more than the subscribers 

```python
# creating a chart using seaborn library to answer this question 
plt.figure(figsize=(10,6))
sb.barplot(x='user_type',y='duration_sec',data=df)
plt.title('Duration time for Subscribers and Customers')
plt.grid()
plt.show()
```

# who uses the bike most? male or female or other?

#### In this question I found what are the gender percentage, and in the following pie chart we found who uses the bike the most out of the 3 genders we have:
#### 1- we can notice that is 75% of the users are male and they were = 130,651 male users
#### 2- also 23% of the users are female users and with the number = 40,844 
#### 3- lastly 2% of the users are other gender and they are = 3652 users 



```python
plt.figure(figsize=(9,9))
df['member_gender'].value_counts(normalize=True).plot(kind='pie',figsize=(10,8), autopct='%1.0f%%')
plt.title('comparing the males and females and others in a pie chart')
plt.legend();
```

```python
df['member_gender'].value_counts(normalize=False).plot(kind='barh',figsize=(10,8),color='orange')
plt.title('comparing the males and females and others in a bar chart')
plt.xlabel('Number of users')
plt.ylabel('Gender')
plt.legend();
```

# The Duration Time vs The Age

#### from the following visualization, what we can notice that the people who are aged from 25 to 35, they spend a lot of time using the bike with nearly 40,000 seconds
#### but in general we can notice most of them are from 20 to 45 but specificlly from 25 to 35 


```python
df['Age']= (2020 - df['member_birth_year'])
plt.figure(figsize=(11,8))
plt.scatter( df['Age'],df['duration_sec'],color='orange')
plt.title('Duration time and age ')
plt.grid()
plt.xlabel('Age')
plt.ylabel('The Seconds on the bike')
plt.show()
```

# Comparing The Age and Time duration with Gender 

#### We can see from the following graph that: 
#### That is the male who spends most of his time on the bike and we can clearly see that from the graph 
#### Also it makes sense why male are the most because they are 75% of the users and it might be related to this graph
#### Also most of the females time spent are in range between 0 to 20,000 seconds 



```python
gender_s = [['Male','a'],['Female','b'],['Other','c']]
```

```python
plt.figure(figsize=(11,8))
for gender, s in gender_s:
    df_g=df[df['member_gender']== gender]
    plt.scatter((df_g['Age']),df_g['duration_sec'])
plt.title('Time duration vs Age vs Gender')
plt.xlabel('Age')
plt.ylabel('Time Duration or Spent')
plt.grid()
plt.legend(['Male','Female','Other']);
```
